```python
%load_ext Cython
import numpy as np
```

```python
def du_dt(u,v):
    return -v

def dv_dt(u,v):
    return u

a,b,u_0,v_0=0,1000,1,0
```

```python
n=250
```

```python
def solve_oed_struc_d2(a, b, n, u_0, v_0):
    dt = (b - a) / n
    u = np.zeros(n+1)
    v = np.zeros(n+1)

    u[0], v[0] = u_0, v_0
    print(u)
    print(v)
    a = np.array([[2,dt],[-dt,2]])
    for i in range(n):
        b = np.dot([[2, -dt], [dt, 2]], [[u[i]], [v[i]]])
        b=b.ravel()
        u[i+1],v[i+1]=np.linalg.solve(a,b)

    return (u,v)
```

```python
u,v=solve_oed_struc_d2(a,b,n,u_0,v_0)
u**2+v**2
```

```python
import matplotlib.pyplot as plt

x=np.linspace(a,b,len(u))
plt.plot(x,u,label="u")
plt.plot(x,v,label="v")
plt.xlabel("t")
plt.legend()
plt.savefig("1SC22317Y05a02.pdf")
plt.show()
```
