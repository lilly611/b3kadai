```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,u):
    return 5*(2-u)*u

a=0
b=1 #closed interval
u_a=0.04
n=100 #division
dt=(b-a)/n #infinitesimal difference
t=[a+i*dt for i in range(n+1)]
u=[0]*n
u[0]=u_a

for i in range(n-1):
    k_1=f(t[i],u[i])
    k_2=f(t[i]+dt/2,u[i]+(1/2)*dt*k_1)
    k_3=f(t[i]+dt/2,u[i]+(1/2)*dt*k_2)
    k_4=f(t[i+1],u[i]+dt*k_3)
    u[i+1]=u[i]+(1/6)*dt*(k_1+2*(k_2+k_3)+k_4)

x_1=[i for i in range(n)]
x_2=np.linspace(0,1,100)
y=[u[i] for i in x_1]
plt.plot(x_1,y)
plt.xticks([0,n],[a,b])
plt.show
```
