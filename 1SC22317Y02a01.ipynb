```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,u):
    return 5*(2-u)*u

a=0
b=1 #closed interval
u_a=0.04
n=100 #division
dt=(b-a)/n #infinitesimal difference
t=[a+i*dt for i in range(n+1)]
u=[0]*n
u[0]=u_a

def solve_oed_euler(a,b,f,n,u_a):
    dt=(b-a)/n
    t=[a+i*dt for i in range(n+1)]
    u=[0]*(n+1)
    u[0]=u_a

    for i in range(n):
        u[i+1]=u[i]+dt*f(t[i],u[i])
    return u

u=solve_oed_euler(a,b,f,n,u_a)
x_1=[i for i in range(n+1)]
x_2=np.linspace(0,1,101)
y=[u[i] for i in x_1]
plt.plot(x_1,y)
plt.xticks([0,n],[a,b])
plt.xlabel("t")
plt.ylabel("u")
plt.show
```
