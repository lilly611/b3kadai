```python
import numpy as np
import matplotlib.pyplot as plt

def f(t,u):
    return t+u

a=0
b=1 #closed interval
u_a=0.5
n=100 #division
dt=(a-b)/n #infinitesimal difference
t=[a+i*dt for i in range(n+1)]
u=[0]*n
u[0]=u_a

for i in range(n-1):
    u[i+1]=u[i]+dt*f(t[i],u[i])

x=[i for i in range(n)]
y=[u[i] for i in x]
plt.plot(x,y)
plt.show
```
